````python
#mesale ghashang ham barat bezanm
import numpy as np
x=np.linspace(0,np.pi , 100)
y=np.sin(x)
#x = [0 , 0.0001 , 0.002,....3.14]
#y = [sin(0) , sin(0.0001) , sin(0.002) , ... sin(3.14)]

plt.plot(x,y,ls=':',c='r')
plt.show()
````

````python
#mesale ghashang ham barat bezanm
import numpy as np
x=np.linspace(0,np.pi , 100)
y=np.sin(x)


plt.plot(x,y,ls=':',c='r')
#ma beyn mitoni ye karaee koni
#2 ta --> title() , xlabel() ylabel()
plt.show()
````

````python
import numpy as np
x=np.linspace(0,np.pi , 100)
y=np.sin(x)


plt.plot(x,y,ls=':',c='r')
plt.title('my sin function')
plt.show()
````

````python
import numpy as np
x=np.linspace(0,np.pi , 100)
y=np.sin(x)


plt.plot(x,y,ls=':',c='r')
plt.title('my sin function')
plt.xlabel('Values')
plt.ylabel('Sin values')
plt.show()
````
