```python
import numpy as np
import matplotlib.pyplot as plt



x = np.linspace(0,np.pi , 50) #200 ta noghte az 0 ta pi
y = np.sin(x) #_->[sin(0) sin(..) ,..... sin(pi)] -->200 ta noghte
#(x,y). (0,0) , (pi/2 , 1) , (pi , 0) ,...

plt.plot(x,y, marker='o' , ms=4, mec='r' , mfc='k')
plt.title('my sinus function')
plt.xlabel('concentration')
plt.ylabel('frequency(Hz)')
plt.show()



```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt


fig = plt.figure()
ax = plt.axes(projection='3d')

z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)

#ax zir majmoeye plt --> advance vidualization
ax.plot3D(x, y, z, 'gray')
plt.show()

```

```python
#nokteye nahaee --> ma fght yek plot ??


x = np.linspace(0,np.pi , 200)
y = np.sin(x)


x_2 = np.linspace(0,np.pi , 200)
y_2 = np.cos(x_2)


plt.plot(x,y, marker='o' , ms=4, mec='r' , mfc='k')
plt.plot(x_2,y_2, marker='x' , ms=4, mec='b' , mfc='y')
plt.show()
```

```python

x = np.linspace(0,np.pi , 200)
y = np.sin(x)


x_2 = np.linspace(0,np.pi , 200)
y_2 = np.cos(x_2)

#scatter, plot --> label
plt.plot(x,y, marker='o' , ms=4, mec='r' , mfc='k' , label='sin(x)')
plt.plot(x_2,y_2, marker='x' , ms=4, mec='b' , mfc='y', label='cos(x)')
plt.show()
```

```python
x = np.linspace(0,np.pi , 200)
y = np.sin(x)


x_2 = np.linspace(0,np.pi , 200)
y_2 = np.cos(x_2)

#scatter, plot --> label
plt.plot(x,y, marker='o' , ms=4, mec='r' , mfc='k' , label='sin(x)')
plt.plot(x_2,y_2, marker='x' , ms=4, mec='b' , mfc='y', label='cos(x)')
plt.legend()
plt.show()
```
